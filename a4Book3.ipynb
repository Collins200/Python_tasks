```python
# libraries
import os
from urllib.request import urlretrieve
```

```python

def download_text(url):
    guten= os.path.basename(url)
    if not os.path.exists(guten):
        urlretrieve(url, guten)
        return guten
```

```python
def vowelCounterbefore(r):
    vowels = 'aeiouyàâæèéêëîïôùûüÿœ'
    v_count = 0
    c_count= 0

    for char in ''.join(r):
        if char in vowels:
            v_count += 1
        elif char != ' ':
            c_count += 1

    return ["Number of vowels before: ", v_count,"Number of consonants before: ", c_count]
```

```python
def simplify(text):
    import unicodedata
    try:
        text = unicode(text, 'utf-8')
    except NameError:
        pass
    text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore').decode("utf-8")
    return str(text)
```
